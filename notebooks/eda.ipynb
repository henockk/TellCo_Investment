```python
import pandas as pd
from sqlalchemy import create_engine
import seaborn as sns
import matplotlib.pyplot as plt

database_name = 'telecom'
table_name= 'xdr_data'

connection_params = { "host": "localhost", "user": "postgres", "password": "00000000",
                    "port": "5432", "database": database_name}

engine = create_engine(f"postgresql+psycopg2://{connection_params['user']}:{connection_params['password']}@{connection_params['host']}:{connection_params['port']}/{connection_params['database']}")

# str or SQLAlchemy Selectable (select or text object)
sql_query = 'SELECT * FROM xdr_data '

df = pd.read_sql(sql_query, con= engine)
```

```python
#Group the data by "Handset Type" and count the occurrences
handset_counts = df.groupby("Handset Type").size()

#Sort the counts in decending order
handset_counts_sorted = handset_counts.sort_values(ascending=False)

#Extract the top 10 handsets
top_10_handsets = handset_counts_sorted.head(10)

print("Top 10 Handsets Used by Customers:")
print(top_10_handsets)
```

```python
#group the data by Handset manufacturers
manufacturers_counts = df.groupby("Handset Manufacturer").size()

#Sort the counts in decending order
manufacturers_counts_sorted = manufacturers_counts.sort_values(ascending=False)

#Extract the top 3 manufacturers
top_3_manufacturers = manufacturers_counts_sorted.head(3)

print("Top 3 manufacturers:")
print(top_3_manufacturers)
```
